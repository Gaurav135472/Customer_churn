```python
positive_words = ['excellent', 'good', 'great', 'satisfied', 'happy', 'fantastic', 'positive', 'amazing', 'love', 'wonderful','Helpful', 'Responsive', 'Friendly', 'Quick', 'Efficient', 'Resolved', 'Polite', 'Courteous', 'Professional', 'Supportive', 'Understanding', 'Knowledgeable', 'Prompt', 'Excellent', 'Effective', 'Satisfied', 'Outstanding', 'Pleasant', 'Reliable']
negative_words = ['bad', 'poor', 'disappointed', 'worst', 'negative', 'hate', 'awful', 'terrible', 'unhappy', 'unsatisfied','Rude', 'Unhelpful', 'Slow', 'Unresponsive', 'Frustrating', 'Ignored', 'Unresolved', 'Incompetent', 'Unprofessional', 'Delayed', 'Difficult', 'Angry', 'Disrespectful', 'Confusing', 'Unclear', 'Inconvenient', 'Terrible', 'Poor', 'Disappointed', 'Unreliable','inefficient', 'irritating', 'neglected', 'overlooked', 'careless', 'dismissive', 'unfriendly', 'uncooperative', 'inconsistent', 'unfair', 'overpriced', 'conflicted','stressful', 'defective', 'unacceptable', 'overcomplicated', 'displeased', 'misleading','unpleasant', 'shoddy']

# Function to classify sentiment based on entities
def extract_sentiment_words(entities):
    if isinstance(entities, str):  # Only process if it's a string
        entities_lower = entities.lower()  # Convert to lowercase for case-insensitive matching
        positive_found = [word for word in positive_words if word in entities_lower]
        negative_found = [word for word in negative_words if word in entities_lower]
        
        # Return the words found, if any, otherwise return 'None'
        if positive_found:
            return f"Positive words: {', '.join(positive_found)}"
        elif negative_found:
            return f"Negative words: {', '.join(negative_found)}"
        else:
            return 'None'
    return 'None'  # Default for non-text entries

# Apply the function to extract the sentiment words
df['sentiment_words'] = df['entities'].astype('str').apply(extract_sentiment_words)
```
